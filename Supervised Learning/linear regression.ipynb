```python
## LINEAR REGRESSION WITH SINGLE VARIABLE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"c:\Users\lenovo\Desktop\DB_progm\dataBase practise\Retail_Sales_Data.csv")
print(df)
```

```python
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.Figure(figsize=(6,4))
plt.title('Retail Sales of furniture', fontdict={'fontsize':10})
plt.scatter(df.Sales, df.Profit, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Sales']],df.Profit)
```

```python
# prediction
reg.predict([[750]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
